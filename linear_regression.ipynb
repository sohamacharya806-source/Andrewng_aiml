Step 1:
Load & Inspect Data

```python
import pandas as pd
import numpy as numpy

url = "https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv"
df = pd.read_csv(url)
df = df.dropna() # removes the value with none or missing values
print(df.head())

```

now prociding with the features

```python
X = df[['median_income']].values
y = df['median_house_value'].values

# one new thing known as feature scaling is used here.

X = (X - X.mean()) / X.std()
y = (y - y.mean()) / y.std()
y = y.reshape(-1, 1)
X = np.c_[np.ones((m, 1)), X]  # add bias
print(df.head())
```

Dry Run Execution Steps 
1. Input Data Y=[2,4,6,8]
2. Calculate Mean (Y.mean()) \
Sum=2+4+6+8=20
Count=4
Mean=20/4=5.0
3. Calculate Standard Deviation (Y.std())
 Using Population Std Dev (N) for simplicity:
Differences from mean: (2-5)=-3,(4-5)=-1,(6-5)=1,(8-5)=3
Squared differences: (9,1,1,9)
Mean of squared diffs (Variance): ((9+1+1+9)/4= 20/4 =5.0)
(StdDev=\sqrt{5.0} = 2.236(approx)) 
4. Apply Formula: (Y - Y.mean()) / Y.std() # Y.std() = 2.236
y(1)= (2-5.0)/2.236= -3/2.236 = -1.34
y(2)=(4-5.0)/2.236= -1/2.236 = -0.45
y(3)=(6-5.0)/2.236= 1/2.236 = 0.45

5. Final Output (Y)

```python
import numpy as np
m, n = X.shape
theta = np.zeros((n, 1))
print("Initial parameters:", theta)
lr = 0.01
epochs = 2000

for _ in range(epochs):
    y_pred = X @ theta
    error = y_pred - y
    gradients = (2/m) * X.T @ error
    theta -= lr * gradients
print("Trained parameters:", theta)
```

Gradient Descent Step-by-Step Calculation Configuration Samples (\(m\)): 2Features (\(n\)):
1 (plus the bias column \(X_{0}=1\))Learning Rate (\(lr\)): 0.1 Data Initialization Inputs (\(X\)): \(\left[\begin{matrix}1&2\\ 1&3\end{matrix}\right]\) (Sample 1: \(x=2\), Sample 2: \(x=3\))Targets (\(y\)): \(\left[\begin{matrix}5\\ 7\end{matrix}\right]\)Initial Weights (\(\theta \)): \(\left[\begin{matrix}0\\ 0\end{matrix}\right]\) 
Step 1: Prediction (\(\^{y}\)) We calculate \(\^{y}_{i}=\theta _{0}X_{i0}+\theta _{1}X_{i1}\) for both samples. Sample 1: \((0\times 1)+(0\times 2)=\mathbf{0}\)Sample 2: \((0\times 1)+(0\times 3)=\mathbf{0}\)Result: \(\^{y}=[0,0]\) 
Step 2: Gradients (\(grads\)) First, calculate the error \((\^{y}-y)\): Error 1: \(0-5=\mathbf{-5}\)Error 2: \(0-7=\mathbf{-7}\) Gradient for \(\theta _{0}\) (Bias): \(\text{Sum}=(Error_{1}\times X_{1,0})+(Error_{2}\times X_{2,0})\rightarrow (-5\times 1)+(-7\times 1)=\mathbf{-12}\)\(grads[0]=\frac{2}{m}\times \text{Sum}\rightarrow \frac{2}{2}\times -12=\mathbf{-12}\)Gradient for \(\theta _{1}\) (Feature \(x\)): \(\text{Sum}=(Error_{1}\times X_{1,1})+(Error_{2}\times X_{2,1})\rightarrow (-5\times 2)+(-7\times 3)=-10-21=\mathbf{-31}\)\(grads[1]=\frac{2}{m}\times \text{Sum}\rightarrow \frac{2}{2}\times -31=\mathbf{-31}\)
Step 3: Weight Update (\(\theta \)) 
Update weights using the formula: \(\theta _{j}=\theta _{j}-(lr\times grads_{j})\) 
Update \(\theta _{0}\): \(0-(0.1\times -12)=\mathbf{1.2}\)
Update \(\theta _{1}\): \(0-(0.1\times -31)=\mathbf{3.1}\) New Weights: \(\theta =\left[\begin{matrix}1.2\\ 3.1\end{matrix}\right]\) 

```python
def mse(X, y, theta):
    return np.mean((X @ theta - y) ** 2)
y_test_pred = X @ theta
rmse = np.sqrt(np.mean((y_test_pred - y) ** 2))
print("RMSE:", rmse)
```

When to Use Normalization
Use normalization when your data does not follow a normal distribution and you have no extreme outliers. 

    Distance-based models: Algorithms like k-Nearest Neighbors (k-NN) and K-Means Clustering perform better because they rely on exact distances between points.
    Neural Networks: Deep learning models often expect inputs in a bounded range (like [0, 1]) to stabilize activation functions.
    Image Processing: Pixel intensities are naturally bounded (e.g., 0–255), making normalization to [0, 1] the standard practice. 

When to Use Standardization
Use standardization as the "default" for many classical machine learning models, especially if your data is approximately bell-shaped. 

    Statistical Models: Algorithms that assume a normal distribution, such as Linear Regression, Logistic Regression, and Linear Discriminant Analysis (LDA).
    Principal Component Analysis (PCA): PCA seeks to maximize variance; standardizing ensures that features with larger original scales do not unfairly dominate the components.
    Support Vector Machines (SVM): Standardizing helps find the optimal separating hyperplane by ensuring feature scales don't distort distance calculations. 

Rule of Thumb

    If your data has outliers, use Standardization (or even Robust Scaling for extreme cases).
    If you are working with Deep Learning or distance-based models and your data is fairly uniform, use Normalization.
    When in doubt, it is common practice to try both and compare model performance using a cross-validation score.
